```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("images.jpeg")
```

```python
plt.imshow(img)
```

```python
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_alt (1).xml")
```

```python
faces = face_detector.detectMultiScale(img, 1.2, 5)
```

```python
for face in faces:
    x,y,w,h = face
    img = cv2.rectangle(img , (x,y) , (x+w, y+h), (255,255,255), 2 )
```

```python
plt.imshow(img)
```

# On WebCam

```python
cam = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cam.read()
    if ret == False:
        continue
        
    faces = face_detector.detectMultiScale(frame, 1.2, 5)
    for face in faces:
        x,y,w,h = face
        frame = cv2.rectangle(frame , (x,y) , (x+w, y+h), (255,255,255), 2 )
        
    cv2.imshow("Face Detection", frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
