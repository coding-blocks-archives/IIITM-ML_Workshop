```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x  = np.arange(5)
```

```python
y  =  x
```

```python
x_squared = x**2
```

```python
plt.plot(x,y, marker = "*" )
plt.plot(x, x_squared, marker = "o")
plt.xlabel("Numbers")
plt.ylabel("values")
plt.show()
```

```python
X = np.random.randn(100,2)
```
